# Regimes de volatilidade 

Histogramas de Volatility Ratio e Volatility CV: Histograma que mostra a distribuição desses indicadores, com uma linha cumulativa (Pareto) sobreposta, que ajuda a visualizar a proporção acumulada dos valores.

Gráficos com Marcadores (Thresholds): Plots de séries temporais do preço de fechamento, com marcadores em pontos em que o Volatility Ratio ou Volatility CV ultrapassam determinados limiares (valores de corte definidos, como 1.216 ou 0.25).

Gráfico de Regressores: Gráficos de dispersão entre as variáveis:

Volatility Ratio x Volatilidade: Relação entre o Volatility Ratio e a volatilidade.
Volatility CV x Volatilidade: Relação entre o Volatility CV e a volatilidade.
Volatility Ratio x Volatility CV: Relação entre Volatility Ratio e Volatility CV, com um ajuste de polinômio de 2° grau para mostrar uma tendência de melhor ajuste.

Esse código realiza uma análise detalhada sobre a volatilidade de um ativo financeiro ao longo do tempo, calcula várias métricas de volatilidade (como o Volatility Ratio e o Volatility CV) e visualiza essas métricas por meio de gráficos interativos, histogramas e regressões, com o intuito de entender melhor o comportamento e as tendências da volatilidade do ativo.

Esses gráficos e métricas podem ser usados para estratégias de trading baseadas na análise da volatilidade e para avaliar riscos associados ao ativo.


```python
# Carregando as bibliotecas
import pandas as pd
import numpy as np

from numpy import mean, absolute
import matplotlib.pyplot as plt
import yfinance as yf
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px
import kaleido

get_ipython().run_line_magic("matplotlib", "inline")
import warnings
warnings.filterwarnings("ignore")
```

```python
# Carregando a base de dados e fazendo os checks iniciais

# Carrega as bases
ticker1 = "BRL=X" 
df1 = yf.download(ticker1, "2015-01-01", progress = False)
df1 = df1.xs(ticker1, level="Ticker", axis=1)

# Parâmetros do algoritmo
periodos = 1

# df1
df1["Returns"] = df1["Close"].pct_change(periodos)*100
df1["Target"] = df1["Returns"].shift(-periodos)
df1["Vol"] = df1["Returns"].rolling(20).std()*np.sqrt(252)
df1.dropna(axis = 0, inplace = True) 

# Calcula a EMA da Vol
n = 20
df1["EMA_Vol"] = df1["Vol"].ewm(span = n, adjust = False).mean()

# Calcula o Volatility Ratio
df1["Volatility_Ratio"] = (df1["Vol"] / df1["EMA_Vol"]).rolling(10).mean() #adiciona um filtro para suavizar

# Calculo do Coeficiente de variação
df1["Volatility_CV"] = (df1["Vol"].rolling(10).std()/df1["EMA_Vol"]).rolling(5).mean()/df1["Volatility_Ratio"] #pondero pelo custo da volatilidade atual

df1.dropna(inplace = True)
```

```python
df1["Volatility_Ratio"].describe()
```

```python
# Criar um histograma
fig = px.histogram(df1, x = "Volatility_Ratio", nbins = 60, color_discrete_sequence = ["#636EFA"], histnorm = "percent")

# Personalizar a aparência
fig.update_layout(
    height = 500, width = 800
    ,title = "Histograma - Volatility Ratio - www.outspokenmarket.com"
    ,xaxis_title = "Value"
    ,yaxis_title = "Percentage"
    ,bargap = 0.01  # Gap between bars
    ,template = "plotly_white"
)

# Plot grafico
fig.show()
```

```python
# Histograma com Pareto
hist_data, bin_edges = np.histogram(df1["Volatility_Ratio"], bins = 40)
bins = bin_edges[:-1]

# Calcular soma cumulativa para a ordem de bin original
cumulative_counts = np.cumsum(hist_data)

# Converter contagens cumulativas em porcentagens
cumulative_percent = (cumulative_counts / cumulative_counts[-1]) * 100

# Plotting
fig = make_subplots(specs = [[{"secondary_y": True}]])

# Barras de histograma
fig.add_trace(
    go.Bar(x = bins, y = hist_data, name = "Count", marker_color = "#636EFA")
    ,secondary_y = False
)

# Linha percentual cumulativa
fig.add_trace(
    go.Scatter(x = bins, y = cumulative_percent, mode = "lines+markers", name = "Cumulative %")
    ,secondary_y = True
)

# Personalizando a aparência
fig.update_layout(
    height = 500, width = 800
    ,title = "Pareto Histogram - Volatility Ratio"
    ,xaxis_title = "Value"
    ,template = "plotly_white"
)
fig.update_yaxes(title_text = "Count", secondary_y = False)
fig.update_yaxes(title_text = "Cumulative %", range = [0, 100], secondary_y = True)

# Plot grafico
fig.show()
```

```python
# Criar subplots
fig = make_subplots(rows = 3, cols = 1, shared_xaxes = True, vertical_spacing = 0.1)

# Plot close price
fig.add_trace(
    go.Scatter(x = df1.index, y = df1["Close"], mode = "lines", name = "Close Price", line = dict(color = "black"))
    ,row = 1, col = 1
)

# Marque os pontos onde a taxa de volatilidade está acima da taxa
ratio_high = 1.216
above_ratio = df1["Volatility_Ratio"] > ratio_high
fig.add_trace(
    go.Scatter(x = df1.index[above_ratio], y = df1["Close"][above_ratio], mode = "markers", marker = dict(color = "red", symbol = "circle"), name = f"Volatility Ratio > {above_ratio}")
    ,row = 1, col = 1
)

# Marque os pontos onde a taxa de volatilidade está abaixo da taxa
ratio_low = 0.72
below_ratio = df1["Volatility_Ratio"] < ratio_low
fig.add_trace(
    go.Scatter(x = df1.index[below_ratio], y = df1["Close"][below_ratio], mode = "markers", marker = dict(color = "blue", symbol = "circle"), name = f"Volatility Ratio < {below_ratio}")
    ,row = 1, col = 1
)

# Taxa de volatilidade do gráfico
fig.add_trace(
    go.Scatter(x = df1.index, y = df1["Volatility_Ratio"], mode = "lines", name = "Volatility Ratio", line = dict(color = "blue"))
    ,row = 2, col = 1
)

# Plot vol
fig.add_trace(
    go.Scatter(x = df1.index, y = df1["Vol"], mode = "lines", name = "Volatility", line = dict(color = "blue"))
    ,row = 3, col = 1
)

# Linha horizontal para limite
fig.add_shape(
    go.layout.Shape(type = "line", line = dict(color = "red", width = 1, dash = "dash"),
                    y0 = ratio_high, y1 = ratio_high, x0 = df1.index.min(), x1 = df1.index.max())
    ,row = 2, col = 1
)

fig.add_shape(
    go.layout.Shape(type = "line", line = dict(color = "red", width = 1, dash = "dash"),
                    y0 = ratio_low, y1 = ratio_low, x0 = df1.index.min(), x1 = df1.index.max())
    ,row = 2, col = 1
)

# Ajustando o layout para os subplots
fig.update_layout(
    height = 1000, width = 800
    ,title = "Volatility Ratio " + str(ticker1) + " www.outspokenmarket.com"
    ,xaxis = dict(title = "Date")
    ,yaxis = dict(title = "Close Price")
    ,yaxis2 = dict(title = "Volatility Ratio")
    ,yaxis3 = dict(title = "Volatility (%)")
    ,showlegend = False
)

fig.update_layout(hovermode = "x unified")

# Código para excluir datas vazias do gráfico
dt_all = pd.date_range(start = df1.index[0]
                       , end = df1.index[-1]
                       , freq = "D")
dt_all_py = [d.to_pydatetime() for d in dt_all]
dt_obs_py = [d.to_pydatetime() for d in df1.index]

dt_breaks = [d for d in dt_all_py if d not in dt_obs_py]

fig.update_xaxes(
    rangebreaks = [dict(values = dt_breaks)]
)

fig.show()
```

```python
# Histograma com Pareto
hist_data, bin_edges = np.histogram(df1["Volatility_CV"], bins = 60)
bins = bin_edges[:-1]

# Calcular soma cumulativa para a ordem de bin original
cumulative_counts = np.cumsum(hist_data)

# Converter contagens cumulativas em porcentagens
cumulative_percent = (cumulative_counts / cumulative_counts[-1]) * 100

# Plotting
fig = make_subplots(specs = [[{"secondary_y": True}]])

# Histograma
fig.add_trace(
    go.Bar(x = bins, y = hist_data, name = "Count", marker_color = "#636EFA")
    ,secondary_y = False
)

# Cumulative percentage line
fig.add_trace(
    go.Scatter(x = bins, y = cumulative_percent, mode = "lines+markers", name = "Cumulative %")
    ,secondary_y = True
)

# Aparencia
fig.update_layout(
    height = 500, width = 800
    ,title = "Pareto Histogram - Volatility CV"
    ,xaxis_title = "Value"
    ,template = "plotly_white"
)
fig.update_yaxes(title_text = "Count", secondary_y = False)
fig.update_yaxes(title_text = "Cumulative %", range = [0, 100], secondary_y = True)

# Plot show
fig.show()
```

```python
# Subplots
fig = make_subplots(rows = 2, cols = 1, shared_xaxes = True, vertical_spacing = 0.1)

# Plot ativo
fig.add_trace(
    go.Scatter(x = df1.index, y = df1["Close"], mode = "lines", name = "Close Price", line = dict(color = "black")),
    row = 1, col = 1
)

# Marque os pontos onde a taxa de volatilidade está acima da taxa
ratio_high = 0.25
above_ratio = df1["Volatility_CV"] > ratio_high
fig.add_trace(
    go.Scatter(x = df1.index[above_ratio], y = df1["Close"][above_ratio], mode = "markers", marker = dict(color = "blue", symbol = "circle"), name = f"Volatility_CV > {above_ratio}")
    ,row = 1, col = 1
)

# Marque os pontos onde o CV de volatilidade está abaixo da proporção
ratio_low = 0.0
below_ratio = df1["Volatility_CV"] < ratio_low
fig.add_trace(
    go.Scatter(x = df1.index[below_ratio], y = df1["Close"][below_ratio], mode = "markers", marker = dict(color = "red", symbol = "circle"), name = f"Volatility_CV < {below_ratio}")
    ,row = 1, col = 1
)

# Plot Volatidade CV
fig.add_trace(
    go.Scatter(x = df1.index, y = df1["Volatility_CV"], mode = "lines", name = "Volatility CV", line = dict(color = "blue"))
   ,row = 2, col = 1
)

# Linha horizontal para limite
fig.add_shape(
    go.layout.Shape(type = "line", line = dict(color = "red", width = 1, dash = "dash"),
                    y0 = ratio_high, y1 = ratio_high, x0 = df1.index.min(), x1 = df1.index.max())
    ,row = 2, col = 1
)

fig.add_shape(
    go.layout.Shape(type = "line", line = dict(color = "red", width = 1, dash = "dash"),
                    y0 = ratio_low, y1 = ratio_low, x0 = df1.index.min(), x1 = df1.index.max())
    ,row = 2, col = 1
)

# Ajustando o layout para os subplots
fig.update_layout(
    height = 800, width = 800
    ,title = "Volatility CV " + str(ticker1) + " www.outspokenmarket.com"
    ,xaxis2 = dict(title = "Date")
    ,yaxis = dict(title = "Close Price")
    ,yaxis2 = dict(title = "Volatility CV")
    ,showlegend = False
)

fig.update_layout(hovermode = "x unified")

# Código para excluir datas vazias do gráfico
dt_all = pd.date_range(start = df1.index[0]
                       , end = df1.index[-1]
                       , freq = "D")
dt_all_py = [d.to_pydatetime() for d in dt_all]
dt_obs_py = [d.to_pydatetime() for d in df1.index]

dt_breaks = [d for d in dt_all_py if d not in dt_obs_py]

fig.update_xaxes(
    rangebreaks = [dict(values = dt_breaks)]
)

fig.show()
```

```python
# Calcular a linha de regressão
m, b = np.polyfit(df1["Volatility_CV"], df1["Vol"], 1)
regression_line = m * df1["Volatility_CV"] + b

# Crie o gráfico de dispersão e a linha de regressão
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = df1["Volatility_CV"],
        y = df1["Vol"],
        mode = "markers",
        name = "Data"
    )
)

fig.add_trace(
    go.Scatter(
        x = df1["Volatility_CV"],
        y = regression_line,
        mode = "lines",
        name = "Regression Line"
    )
)

# Defina títulos e rótulos
fig.update_layout(
    height = 600, width = 800,
    title = "Volatility CV x Vol " + str(ticker1) + " www.outspokenmarket.com",
    xaxis_title = "Volatility_CV",
    yaxis_title = "Vol"
)

# Plot
fig.show()
```

```python
# Calcular a linha de regressão
m, b = np.polyfit(df1["Volatility_Ratio"], df1["Vol"], 1)
regression_line = m * df1["Volatility_Ratio"] + b

# Crie o gráfico de dispersão e a linha de regressão
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = df1["Volatility_Ratio"],
        y = df1["Vol"],
        mode = "markers",
        name = "Data"
    )
)

fig.add_trace(
    go.Scatter(
        x = df1["Volatility_Ratio"],
        y = regression_line,
        mode = "lines",
        name = "Regression Line"
    )
)

# SDefina títulos e rótulos
fig.update_layout(
    height = 600, width = 800,
    title = "Volatility Ratio x Vol " + str(ticker1) + " www.outspokenmarket.com",
    xaxis_title = "Volatility_Ratio",
    yaxis_title = "Vol (%)"
)

# Plot
fig.show()
```

```python
# Calcular o ajuste polinomial de segundo grau
coefficients = np.polyfit(df1["Volatility_Ratio"], df1["Volatility_CV"], 2)
polynomial_vals = np.polyval(coefficients, df1["Volatility_Ratio"])

# Crie o gráfico de dispersão e a linha de regressão
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = df1["Volatility_Ratio"],
        y = df1["Volatility_CV"],
        mode = "markers",
        name = "Data"
    )
)

fig.add_trace(
    go.Scatter(
        x = df1["Volatility_Ratio"],
        y = polynomial_vals,
        mode = "lines",
        name = "2nd degree polinomial"
    )
)

# Defina títulos e rótulos
fig.update_layout(
    height = 600, width = 800,
    title = "Volatility Ratio x Volatility_CV " + str(ticker1) + " www.outspokenmarket.com",
    xaxis_title = "Volatility_CV",
    yaxis_title = "Volatility Ratio"
)

# Plot
fig.show()
```
